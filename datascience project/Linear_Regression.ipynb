# classification - Linear Regression

```python
# Dividing input and output labels
X = df.drop(['output'],axis=1)
y = df['output']
```

```python
# importing Min Max scalar for the data transformation
# Scaling is very important in the case of linear regression
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_X = scaler.fit_transform(X)
print(scaled_X)
```

```python
# dividing the dataset into train,test and validation sets
X_train, X_rem, y_train, y_rem  = train_test_split(scaled_X, y, train_size = 0.5, random_state = 100)
X_valid, X_test, y_valid, y_test = train_test_split(X_rem, y_rem, test_size = 0.5, random_state = 100)
```

```python
# The shape of training dataset
print(f" The input training  data shape is {X_train.shape}")
print(f" The  output training data shape is {y_train.shape}")
print(f" The input validation  data shape is {X_valid.shape}")
print(f" The  output validation data shape is {y_valid.shape}")
# The shape of the test dataset 
print(f" The input testing  data shape is {X_valid.shape}")
print(f" The  output testing data shape is {y_valid.shape}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report, recall_score, accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
```

```python
# finding the best model using the gridsearch
from sklearn.model_selection import GridSearchCV
param_grid = {'fit_intercept': [True, False], 'copy_X': [True, False]}
```

```python
models = []
for c in np.unique(y):
    y_binary_train = (y_train == c).astype(int)
    y_binary_val = (y_valid == c).astype(int)
    model = GridSearchCV(LinearRegression(), param_grid, scoring='accuracy', cv=5).fit(X_train, y_binary_train 
                                                                                       )
    models.append(model.best_estimator_)

filename = "Linear_class.sav"
pickle.dump(models, open(filename, 'wb'))
```

### Roc auc score for the models 

```python
models = []
for c in np.unique(y):
    y_binary_train = (y_train == c).astype(int)
    y_binary_val = (y_valid == c).astype(int)
    model = GridSearchCV(LinearRegression(), param_grid, scoring='accuracy', cv=5).fit(X_train, y_binary_train)
    models.append(model.best_estimator_)
# compute ROC AUC score for training data
train_predictions = np.zeros((X_train.shape[0], len(models)))
for i, model in enumerate(models):
    train_predictions[:, i] = model.predict(X_train)
train_roc_auc_scores = []
for c in np.unique(y):
    y_binary_train = (y_train == c).astype(int)
    if c<=i:
      y_pred = train_predictions[:, c]
      roc_auc = roc_auc_score(y_binary_train, y_pred)
      train_roc_auc_scores.append(roc_auc)
train_avg_roc_auc_score = np.mean(train_roc_auc_scores)
```

```python

# compute ROC AUC score for validation data
val_predictions = np.zeros((X_valid.shape[0], len(models)))
for i, model in enumerate(models):
    val_predictions[:, i] = model.predict(X_valid)
val_roc_auc_scores = []
for c in np.unique(y):
    y_binary_val = (y_valid == c).astype(int)
    if c <= i:
       y_pred = val_predictions[:, c]
       roc_auc = roc_auc_score(y_binary_val, y_pred)
       val_roc_auc_scores.append(roc_auc)
val_avg_roc_auc_score = np.mean(val_roc_auc_scores)
```

```python
# compute ROC AUC score for testing data
test_predictions = np.zeros((X_test.shape[0], len(models)))
for i, model in enumerate(models):
    test_predictions[:, i] = model.predict(X_test)
test_roc_auc_scores = []
for c in np.unique(y):
    y_binary_test = (y_test == c).astype(int)
    if c <= i:
       y_pred = test_predictions[:, c]
       roc_auc = roc_auc_score(y_binary_test, y_pred)
       test_roc_auc_scores.append(roc_auc)
test_avg_roc_auc_score = np.mean(test_roc_auc_scores)
```

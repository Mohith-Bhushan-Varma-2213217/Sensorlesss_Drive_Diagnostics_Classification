```python
#import requuired libraries
import pandas as pd
import numpy as np
import sys
import joblib
sys.modules['sklearn.externals.joblib'] = joblib
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV , train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, roc_auc_score, recall_score

import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Group_9_data_cleaned.csv')
df
```

```python
df = df.iloc[:,1:]
df
```

```python
X = df.drop(['output'],axis=1)
y = df['output']
```

```python
X_train, X_test, y_train, y_test  = train_test_split(X, y, train_size = 0.5, random_state = 42)
```

# Random Forest

```python
randf = RandomForestClassifier()
```

```python
sfs = SFS(randf,
         k_features=(1,49),forward=True,
                                floating=True,
                                scoring='accuracy')
```

```python
sfs.fit(X_train,y_train)
```

```python
sfs.k_feature_names_
```

```python
x_train_feat = X_train[list(sfs.k_feature_names_)]
x_test_feat=X_test[list(sfs.k_feature_names_)]
```

```python
# Define a grid of hyperparameters to search over
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [5, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
}
```

```python
# Create a Grid Search CV object and fit it to the Validation data
grid_search = GridSearchCV(randf, param_grid, cv=5)
grid_search.fit(x_train_feat, y_train)
```

```python
grid_search.best_params_
```

```python
# best classification tree for the above data
best_rf = RandomForestClassifier(max_depth = 10, min_samples_leaf= 1, min_samples_split= 2, n_estimators= 50)
best_rf.fit(x_train_feat,y_train)
filename = 'RF_class.sav'
pickle.dump(best_rf, open(filename, 'wb'))
```

```python
# Best model training 
best_rf.fit(x_train_feat,y_train)
# Accuracy scores of best tree on different datasets
y_train_pred = best_rf.predict(x_train_feat)
print(f" The training data accuracy for the best model is {accuracy_score(y_train,y_train_pred)}")
y_test_pred = best_rf.predict(x_test_feat)
print(f" The testing data accuracy for the best model is {accuracy_score(y_test,y_test_pred)}") 
```

```python
### Training Data
print(confusion_matrix(y_train,y_train_pred))
```

```python
# Confusion matrix for testing data
print(confusion_matrix(y_test,y_test_pred))
```

```python
#Classification report for the training data
print(classification_report(y_train,y_train_pred))
```

```python
# Classification report for the testing data
print(classification_report(y_test,y_test_pred))
```

```python
# roc auc score for training data
y_train_prob = best_rf.predict_proba(x_train_feat)
print(f" The roc_auc score for the training data is {roc_auc_score(y_train,y_train_prob,average='weighted',multi_class='ovr')}")
# roc auc score for the testing data
y_test_prob = best_rf.predict_proba(x_test_feat)
print(f" The roc_auc score for the testing data is {roc_auc_score(y_test,y_test_prob,average='weighted',multi_class='ovr')}")
```

```python
# recall score for training data
print(f" The recall score for the training data is {recall_score(y_train,y_train_pred,average= 'macro')}")
# recall score for test data
print(f" The recall score for the testing data is {recall_score(y_test,y_test_pred,average= 'macro')}")
```

# Using Deep Learning


```python
# Importing the deep learning model
from sklearn.neural_network import MLPClassifier
```

```python
from sklearn.preprocessing import MinMaxScaler
X_train_scaled=df
scaler = MinMaxScaler()
X_train_scaled.iloc[:,0:49] = scaler.fit_transform(X_train_scaled.iloc[:,0:49])
X_train_scaled=X_train_scaled.drop(['output'],axis=1)
X_train_scaled
```

```python
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled  = train_test_split(X_train_scaled, y, train_size = 0.5, random_state = 42)
```

```python
from sklearn.model_selection import RandomizedSearchCV
mlp = MLPClassifier()
```

```python
sfs_deep = SFS(mlp,
         k_features=(1,49),forward=True,
                                floating=True,
                                scoring='accuracy')
```

```python
sfs_deep.fit(X_train_scaled,y_train_scaled)
```

```python
#best fetures
sfs_deep.k_feature_names_
```

```python
x_train_deep_feat = X_train_scaled[list(sfs_deep.k_feature_names_)]
x_test_deep_feat=X_test_scaled[list(sfs_deep.k_feature_names_)]
```

```python
# Define the parameter grid
param_grid = {
    'hidden_layer_sizes': [(100,), (200,), (500,), (1000,)],
    'activation': ['relu', 'logistic'],
    'solver': ['adam', 'sgd'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate': ['constant', 'adaptive'],
}
```

```python
# Perform grid search cross-validation
clf = RandomizedSearchCV(mlp, param_grid, random_state=0)
```

```python
clf.fit(x_train_deep_feat, y_train_scaled)
```

```python
# Print the best hyperparameters
print(f'Best hyperparameters: {clf.best_params_}')
```

```python
# best model 
best_model_mlp =  MLPClassifier(solver= 'adam', learning_rate= 'adaptive', hidden_layer_sizes= (500,), alpha= 0.001, activation= 'relu')
best_model_mlp.fit(x_train_deep_feat,y_train_scaled)
```

```python
filename = 'mlp_class.sav'
pickle.dump(best_model_mlp, open(filename, 'wb'))
```

```python
y_train_pred_deep_feat = best_model_mlp.predict(x_train_deep_feat)
y_test_pred_deep_feat  = best_model_mlp.predict(x_test_deep_feat)
print(f" The training data accuracy for the best model is {accuracy_score(y_train_scaled,y_train_pred_deep_feat)}")
print(f" The testing data accuracy for the best model is {accuracy_score(y_test_scaled,y_test_pred_deep_feat)}")
```

```python
### Training Data
print(confusion_matrix(y_train_scaled,y_train_pred_deep_feat))
```

```python
### Testing Data
print(confusion_matrix(y_test_scaled,y_test_pred_deep_feat))
```

```python
#Classification report for the training data
print(classification_report(y_train_scaled,y_train_pred_deep_feat))
```

```python
#Classification report for the testing data
print(classification_report(y_test_scaled,y_test_pred_deep_feat))
```

```python
# roc auc score for training data
y_train_prob = best_model_mlp.predict_proba(x_train_deep_feat)
print(f" The roc_auc score for the training data is {roc_auc_score(y_train_scaled,y_train_prob,average='weighted',multi_class='ovr')}")
# roc auc score for the testing data
y_test_prob = best_model_mlp.predict_proba(x_test_deep_feat)
print(f" The roc_auc score for the testing data is {roc_auc_score(y_test_scaled,y_test_prob,average='weighted',multi_class='ovr')}")
```

```python
# recall score for training data
print(f" The recall score for the training data is {recall_score(y_train_scaled,y_train_pred_deep_feat,average= 'macro')}")
# recall score for test data
print(f" The recall score for the testing data is {recall_score(y_test_scaled,y_test_pred_deep_feat,average= 'macro')}")
```

```python
# F-1 score for training data
print(f" The f-1 score for the training data is {f1_score(y_train_scaled,y_train_pred_deep_feat,average= 'weighted')}")
# F-1 score for test data
print(f" The f-1 score for the testing data is {f1_score(y_test_scaled,y_test_pred_deep_feat,average= 'weighted')}")
```
